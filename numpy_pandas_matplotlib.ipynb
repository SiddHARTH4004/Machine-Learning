# [1] Use of numpy array:

#### (1) print array , data type , size of the array & print sorted array : 

```python
import numpy as np

arr=np.array([6,9,2,5])
print("number: ",arr)
print("data type of array: ",arr.dtype)
print("size: ",np.size(arr))
print("sorted array: ",np.sort(arr))
```

### (2) addition & multiplication of numpy array :

```python
arr_1=np.array([5,3,1])
arr_2=np.array([6,4,2])
print("addition of 2 array:",arr_1+arr_2)
print("multiplication of 2 array:",arr_1*arr_2)
```

# [3] use of matplotlib :

### (1) simple line plot :

```python
import matplotlib.pyplot as plt 
x=[1,2,3]
y=[10,20,30]
plt.plot(x,y)

plt.title("using matplotlib: Line Plot",color='blue', fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### (2) scatter plot:

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[10,20,30]
plt.scatter(x,y,marker='+')

plt.title(" Scatter plot", color='blue',fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### (3) bar plot:

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[10,20,30]
plt.bar(x,y)

plt.title("Bar plot",color='blue',fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### (4) Customizing Line Plot :

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[10,20,30]
plt.plot(x,y,color='red',linestyle='--',marker='^',markerfacecolor='yellow',markersize=15,linewidth='1')

plt.title("Line plot",color='blue',fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Using Matplotlib with pandas :

```python
import pandas as pd 
import matplotlib.pyplot as plt
data={ 
      "subject" :["os","ml","java"] ,
      "SE1":[20,23,25]
     } 
df=pd.DataFrame(data)
print(df)

df.plot(x='subject',y='SE1',kind='line',linestyle='--',linewidth='1',marker='*',markerfacecolor='yellow',markersize='15')

plt.title("SE1 marks")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show()
```

### change color, fontsize, fontweight:

```python
import pandas as pd 
import matplotlib.pyplot as plt
data={ 
      "subject" :["os","ml","java"] ,
      "SE1":[20,23,25]
     } 
df=pd.DataFrame(data)
print(df)

df.plot(x='subject',y='SE1',kind='line',linestyle='--',linewidth='1',marker='*',markerfacecolor='yellow',markersize='15')

plt.title("SE1 marks",fontsize=20,color='orange',fontweight='bold')
plt.xlabel("subject",fontsize=20,color='blue')
plt.ylabel("marks",fontsize=20,color='blue')

plt.xticks(fontsize=20)
plt.yticks(fontsize=15)


plt.show()
```
